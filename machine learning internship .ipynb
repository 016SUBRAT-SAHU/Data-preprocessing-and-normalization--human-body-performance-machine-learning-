# Decimal scaling normalization


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import FunctionTransformer
```

```python
data={"bedroom0":[1,2,3,4,5],
      "barthroom":[1,2,2,3,2],
      "sqliving":[500,1000,1500,2000,2500],
      "age of house":[5,10,15,20,25],
      "price":[100000,150000,200000,250000,300000]}
```

```python
df=pd.DataFrame(data)
df
```

**creat the decimal_scaler library by our owndataset**

```python
def decimal_scaler(x):
  max_abv=np.max(np.abs(x))
  scale=10**np.ceil(np.log10(max_abv))
  return x/scale
```

**applying the decimal_scaling techin**

```python
transformer=FunctionTransformer(decimal_scaler)
```

*fit the data*

```python
decimal_scaling=transformer.fit_transform(df)
decimal_scaling
```
